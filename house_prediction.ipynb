```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
data = fetch_california_housing()
house_data = pd.DataFrame(data.data, columns = data.feature_names)
house_data.head()
```

```python
house_data['Target_price'] = data.target
house_data.head()
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split

X = house_data[['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population', 'AveOccup', 'Latitude', 'Longitude']]
y = house_data[['Target_price']]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.8, random_state=42)


lr = LinearRegression()

lr.fit(X_train,y_train)

y_pred = lr.predict(X_test)

```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, y_pred)

print(mse)
```

```python
import matplotlib.pyplot as plt
import numpy as np

y_test_arr = np.array(y_test[:10])
y_pred_arr = np.array(y_pred[:10])

plt.figure(figsize=(15,10))
plt.plot(y_test_arr, label = 'actual', color = 'red')
plt.plot(y_pred_arr, label = 'predicted', color = 'blue')

plt.ylabel('Target_values')

plt.show()
```
